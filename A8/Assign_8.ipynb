```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import numpy as np

dt = sns.load_dataset('titanic')
dt.head()
```

```python
# Distplot visualization

# displt = sns.distplot(x=data['fare'], bins=10)

displt = sns.distplot(x=dt['fare'], bins=20, color='red', kde=False)

displt
```

```python
# Joint Plot Visualization

jtpt = sns.jointplot( x = dt['fare'], y = dt['pclass'], kind='scatter')
```

```python
# Rug plot

rugplt = sns.rugplot(dt['fare'])
```

<center>
    <p>Categorical Plot</p>

</center>

```python
# Bar plot

bar = sns.barplot(x = 'fare', y = 'pclass', data=dt)
```

```python
# Bar plot

bar = sns.barplot(x = 'sex', y = 'fare', data=dt, estimator=np.std)
```

```python
# Count plot

ctplt = sns.countplot(x = 'fare', data=dt)
```

```python
# Box plot
# bxplt = sns.boxplot(x = 'sex', y= 'fare', data=dt)


bxplt = sns.boxplot(x = 'sex', y= 'fare', data=dt, hue='survived')
```

```python
# Violin Plot

vplt = sns.violinplot(x = 'sex', y = 'fare', data=dt)
```

```python
# Violin Plot vwith Hue


vplt = sns.violinplot(x = 'sex', y = 'fare', data=dt, hue='survived')
```

<center>
    <p>Advanced Plot</p>

</center>

```python
#  Strip Plot

splt = sns.stripplot( x = 'sex', y = 'fare', data=dt, jitter=True)
```

```python
#  Strip Plot with Hue

splt = sns.stripplot( x = 'sex', y = 'fare', data=dt, jitter=True, hue='survived')
```

```python
# Swarm Plot

# swplt = sns.swarmplot( x = 'sex', y = 'fare', data=dt, hue='survived')
swplt = sns.swarmplot( x = 'sex', y = 'fare', data=dt)
```
